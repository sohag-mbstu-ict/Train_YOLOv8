```python

img_rgb = cv2.imread('/Template/MyDrive/template_matching/images/frame_0003.jpeg')
img_gray = cv2.cvtColor(img_rgb, cv2.COLOR_BGR2GRAY)
template = cv2.imread('/Template/MyDrive/template_matching/images/template.png', 0)
h, w = template.shape[::]

#methods available: ['cv2.TM_CCOEFF', 'cv2.TM_CCOEFF_NORMED', 'cv2.TM_CCORR',
#            'cv2.TM_CCORR_NORMED', 'cv2.TM_SQDIFF', 'cv2.TM_SQDIFF_NORMED']
```

```python

res = cv2.matchTemplate(img_gray, template, cv2.TM_SQDIFF)
# For TM_SQDIFF, Good match yields minimum value; bad match yields large values
# For all others it is exactly opposite, max value = good fit.
plt.imshow(res, cmap='gray')
```

```python

min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)

top_left = min_loc  #Change to max_loc for all except for TM_SQDIFF
bottom_right = (top_left[0] + w, top_left[1] + h)
cv2.rectangle(img_rgb, top_left, bottom_right, 255, 2)  #White rectangle with thickness 2.

cv2_imshow(img_rgb)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python

import cv2
import numpy as np

# Load input image and convert to grayscale
image = cv2.imread('/Template/MyDrive/template_matching/images/frame_0003.jpeg')
# cv2_imshow(image)
cv2.waitKey(0)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Load Template image
template = cv2.imread('/Template/MyDrive/template_matching/images/template.png',0)

result = cv2.matchTemplate(gray, template, cv2.TM_CCOEFF)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(result)

#Create Bounding Box
top_left = max_loc
bottom_right = (top_left[0] + 50, top_left[1] + 50)
cv2.rectangle(image, top_left, bottom_right, (0,0,255), 2)

cv2_imshow(image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
